```python
import requests
from urllib.parse import urlparse
import pandas as pd
```

```python
# 키워드 검색 함수
def keyword_search(keyword):
    url = "https://dapi.kakao.com/v2/local/search/keyword.json?&query=" + keyword
    result = requests.get(urlparse(url).geturl(),
                          headers={"Authorization":"KakaoAK ca3e9e2867f5f89b6efcd2c1c3624ab3"})
    json_obj = result.json()
    list = []
    for document in json_obj['documents']:
        val = [keyword, document['place_name'],
               document['address_name'],
               document['category_name'],
               document['category_group_code'],
               document['category_group_name'],
               document['y'],
               document['x']]
        list.append(val)
    df = pd.DataFrame(list, columns = ['keyword', 'place_name', 'address_name','category_name',
                                       'category_group_code','category_group_name','lat','lon'])
    return df
```

```python
df = pd.read_csv('C:/Users/user/Desktop/문정원/data/category.csv',encoding='cp949')
df.head()
```

```python
df1 = pd.DataFrame(columns = ['keyword', 'place_name', 'address_name','category_name','category_group_code','category_group_name', 
                              'lat', 'lon'])
# 키워드 검색 함수 반복 적용
for keyword in df['카카오위치명']:
    df1 = pd.concat([df1, keyword_search(keyword)])
df1.head()
```
