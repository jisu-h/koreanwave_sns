```python
import pandas as pd
import numpy as np
import folium
from folium import Marker
from folium.plugins import MarkerCluster
```

```python
df = pd.read_csv('C:/Users/user/Desktop/문정원/koreagram_2021_2022_final.csv',encoding='cp949')
df.head()
```

```python
df2 = df[df['대상여부']== 2]
df2.head()
```

## 서울을 중심으로 위치 시각화

```python
# 중심 지정
lat = 37.550
long =126.982

m = folium.Map([lat, long], zoom_start=11)

# 지도위에 표시
for i in df2.index:
    sub_lat = df2.loc[i, '위도']
    sub_long = df2.loc[i, '경도']
    
    title = df2.loc[i, '카카오위치명']
    
    #지도에 데이터 찍어서 보여주기
    folium.Marker([sub_lat, sub_long],tooltip=title,
    icon=folium.Icon(
        color='blue',
        icon_color='white',
        icon='glyphicon glyphicon-thumbs-up',
        prefix='glyphicon'
    )
                 ).add_to(m)

#m.save('seoul_mark.html')
#m
```

## markercluster로 시각화하기

```python
# 지도 생성하기
m = folium.Map(location=[37.5536067,126.9674308],
               zoom_start=12)
marker_cluster = MarkerCluster().add_to(m)  # 클러스터 추가하기

#마커 추가하기

for i in df2.index:
    lat = df2.loc[i, '위도']
    long = df2.loc[i, '경도']
    name = df2.loc[i,'카카오위치명']
    num = df2.loc[i, 'count']


    folium.Marker([lat, long],tooltip= name, popup='%d개'%num).add_to(marker_cluster)    

#m.save('korea_markercluster.html')
#m
```
