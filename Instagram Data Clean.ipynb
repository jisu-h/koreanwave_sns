```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('./southkoreatrip_insta_crawling.csv', encoding='utf-8')
```

```python
df['content']
```

```python
# 정규표현식 사용 # 제외한 특수문자 제거
df['content']= df['content'].str.replace(pat=r'[^A-Za-z0-9가-힣#]',repl=r' ',regex=True)
df['content']
```

```python
# 중복제거
df.drop_duplicates(subset=['content'] , inplace=True)
df = df.reset_index(drop=True)
df
```

```python
# 대상여부 컬럼 추가
df['대상여부'] = 'Y'
df
```

```python
# 한글이 포함되어 있으면 'N'
korean = re.compile(r'[ㄱ-ㅣ가-힣]')
for i in range(len(df['content'])):
    if korean.match(df['content'][i]):
        df['대상여부'][i] = 'N'
```

```python
df['대상여부'].value_counts()
```

```python
df.to_csv('./data/koreagram_2021_2022_clean.csv', index=False, encoding='utf-8-sig')
```
